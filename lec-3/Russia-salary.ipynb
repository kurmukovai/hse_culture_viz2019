# Считывание и чистка данных

```python
salary = pd.read_excel('salary-2013-2018.xlsx', skiprows=2, header=2)
salary.head(2)
```

```python
salary = salary.transpose()
salary.head(2)
```

```python
salary.columns = salary.iloc[0]
salary.head(2)
```

```python
salary = salary.drop('Unnamed: 0', axis=0)
salary.head(2)
```

```python
salary['Месяц'] = salary.index
salary.tail()
```

```python
salary['Месяц'] = [month.split('.')[0] for month in salary['Месяц']]
salary.tail()
```

```python
salary['Год'] = [2013]*12 + [2014]*12 + [2015]*12 + [2016]*12 + [2017]*12 + [2018]*8
salary.head(2)
```

```python
salary['Год_Месяц'] = ['_'.join([str(year), month]) for year, month in zip(salary['Год'], salary['Месяц'])]
salary.head()
```

```python
salary.index = np.arange(salary.shape[0])
salary.head(2)
```

```python
salary.columns
```

```python
salary = salary.drop(' 1) Южный федеральный округ без учета данных по Республике Крым и г. Севастополю', axis=1)
```

```python
salary.drop(np.nan, axis=1, inplace=True)
```

```python
salary.columns
```

```python
salary
```

```python
print(salary.columns.name)
```

```python
salary.groupby('Месяц')['Российская Федерация '].mean()
```

```python
columns = salary.columns[:-3]
columns
```

```python
salary[columns] = salary[columns].astype(float)
```

```python
m2num = {
   'январь': 1,
   'февраль': 2,
   'март': 3,
   'апрель': 4,
   'май': 5,
   'июнь': 6,
   'июль': 7,
   'август': 8,
   'сентябрь': 9,
   'октябрь': 10,
   'ноябрь': 11,
   'декабрь': 12, 
}
salary['Номер Месяца'] = salary['Месяц'].apply(lambda x: m2num[x])
```
